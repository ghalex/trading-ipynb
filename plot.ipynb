```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
counter = Counter()

df = pd.read_csv('mock/languages.csv')
df.rename(columns={"Responder_id": "Id", "LanguagesWorkedWith": "Languages"}, inplace=True)

df['Languages'].apply(lambda x: counter.update(x.split(';')))

df2 = pd.DataFrame.from_dict(counter, orient='index', columns=['Count'])
df2.sort_values(by='Count', ascending=True, inplace=True)
df2.reset_index(inplace=True, names=['Language'])
df2.plot(kind='barh', x='Language', y='Count', figsize=(30, 10), title='Most Popular Languages')
```

```python
text_kwargs = dict(ha='center', va='center', fontsize=28, color='C1')

px = 1/plt.rcParams['figure.dpi']  # pixel in inches
plt.subplots(figsize=(600*px, 200*px))
plt.text(0.5, 0.5, '600px x 200px', **text_kwargs)
```
